**WHAT IS THE OVERALL SALES TREND?**

Getting month year from order_date

```python
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

Grouping month_year by sales

```python
df_temp = df.groupby('month_year')[['sales']].sum().reset_index()
```

Line graph of month_year and sales

```python
# Setting the figure size
plt.figure(figsize=(16, 5))
plt.plot(df_temp['month_year'], df_temp['sales'], color='#b80045')
plt.xticks(rotation='vertical', size=8)
plt.show()
```

**WHICH ARE THE TOP TEN PRODUCTS BY SALES?**

First, group the prodicts by sales.
Then, sort the dataframe in descending order.
Finally, display the top 10 products by sale.

```python
prod_sales = pd.DataFrame(df.groupby('product_name')[['sales']].sum())
prod_sales.sort_values(by=['sales'], inplace=True, ascending=False)
prod_sales[:10]
```

**WHICH ARE THE MOST SELLING PRODUCTS**

```python
best_selling_prod = df.groupby('product_name')[['quantity']].sum()
best_selling_prod.sort_values(by=['quantity'], inplace=True, ascending=False)
best_selling_prod[:10]
```

**WHAT IS THE MOST PREFERRED SHIP MODE?**

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='ship_mode', data=df)
plt.show()
```

By this we can easily see that the most preferred ship mode is Standard Class.

**WHICH ARE THE MOST PROFITABLE CATEGORY AND SUB-CATEGORY?**

```python
df_profitcatsubcat = pd.DataFrame(df.groupby(['category','sub_category'])[['profit']].sum())
df_profitcatsubcat.sort_values(by=['category','profit'], inplace=True, ascending=False)
df_profitcatsubcat
```

## **DATA VISUALISATIONS**
1. Sales Trend over Time
2. Top-Performing Product Categories and Sub-categories
3. Sales, Profit, and Discount Relationship

```python
# 1. Sales Trend over Time
plt.figure(figsize=(12, 6))
sales_by_date = df.groupby('order_date')['sales'].sum()
sns.lineplot(x=sales_by_date.index, y=sales_by_date.values)
plt.xlabel("Order Date")
plt.ylabel("Total Sales")
plt.title("Sales Trend over Time")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Top-Performing Product Categories and Sub-categories
category_sales_profit = df.groupby('category').agg({'sales': 'sum', 'profit': 'sum'})
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.barplot(x=category_sales_profit.index, y=category_sales_profit['sales'])
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.title("Sales by Category")
plt.xticks(rotation=45)
plt.subplot(1, 2, 2)
sns.barplot(x=category_sales_profit.index, y=category_sales_profit['profit'])
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.title("Profit by Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

subcategory_sales_profit = df.groupby('sub_category').agg({'sales': 'sum', 'profit': 'sum'})
plt.figure(figsize=(10, 5))
subcategory_sales_profit = subcategory_sales_profit.sort_values(by='sales', ascending=False)
plt.subplot(1, 2, 1)
sns.barplot(x=subcategory_sales_profit.index[:10], y=subcategory_sales_profit['sales'][:10])
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")
plt.title("Top 10 Sales by Sub-Category")
plt.xticks(rotation=90)

plt.subplot(1, 2, 2)
subcategory_sales_profit = subcategory_sales_profit.sort_values(by='profit', ascending=False)
sns.barplot(x=subcategory_sales_profit.index[:10], y=subcategory_sales_profit['profit'][:10])
plt.xlabel("Sub-Category")
plt.ylabel("Total Profit")
plt.title("Top 10 Profit by Sub-Category")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# 3. Sales, Profit, and Discount Relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sales', y='profit', hue='discount')
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales vs. Profit (Colored by Discount)")
plt.show()
```
